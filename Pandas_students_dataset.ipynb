```python
# Exercise 3 Question 1 - "Write the code to add a new column 'passed' that indicates whether the student passed (mark >= 60)."

dataframe['passed'] = dataframe['mark'] >= 60
print(dataframe)
```

```python
# Exercise 3 Question 2 - "Write the code to rename the 'mark' column to 'score'."

dataframe.rename(columns={'mark': 'score'}, inplace = True)
print(dataframe)
```

```python
# Exercise 3 Question 3 - "Write the code to drop the 'passed' column."

dataframe.drop(columns=['passed'])

# 'mark' column has been renamed to 'score'.
```

```python
# Exercise 4 Question 1- "Write the code to group the DataFrame by the 'class' column and calculate the mean 'mark' for each group."

dataframe.groupby('class')['score'].mean()

# 'mark' column has been renamed to 'score'.
```

```python
# Exercise 4 Question 2 - "Write the code to count the number of students in each class."

dataframe['class'].value_counts()
```

```python
# Exercise 4 Question 3 - "Write the code to calculate the average mark for each gender."

dataframe.groupby('gender')['score'].mean()

# 'mark' column has been renamed to 'score'.
```

```python
# Exercise 5 Question 1 - "Write the code to create a pivot table with 'class' as rows, 'gender' as columns, and 'mark' as values."

dataframe.pivot_table(index='class',columns='gender', values='score')

# 'mark' was renamed to 'score', so we use 'score' as the values.
```

```python
#Exercise 5 Question 2 - "Write the code to create a new column 'grade' where marks >= 85 are 'A', 70-84 are 'B', 60-69 are 'C', and below 60 are 'D'.

dataframe['grade'] = pd.cut(
    dataframe['score'],
    bins=[0,60,70,85,100],
    labels=['D','C','B','A']
)
print(dataframe)

# 'mark' column has been renamed to 'score'.
```

```python
# Exercise 5 Question 3 - "Write the code to sort the DataFrame by 'mark' in descending order."

dataframe.sort_values(by='score', ascending=False)

# 'mark' column has been renamed to 'score'.
```

```python
# Exercise 6 Question 1 - "Write the code to save the DataFrame with the new 'grade' column to a new CSV file."

dataframe.to_csv('student_grades.csv')
```

```python
# Exercise 7 - Plotting Graphs

# Plot 1: Distribution of Scores Across Genders

# Used Output for Exercise 3 Question 3 to plot the graph

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))

#Renaming the chart and axis titles.
  xs = series['id']
  ys = series['score']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = dataframe.sort_values('id', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('gender')):
  _plot_series(series, series_name, i)
  fig.legend(title='gender', bbox_to_anchor=(1, 1), loc='upper left')
  plt.title('Score Comparison by Gender')
sns.despine(fig=fig, ax=ax)
plt.xlabel('id')
_ = plt.ylabel('score')
```
